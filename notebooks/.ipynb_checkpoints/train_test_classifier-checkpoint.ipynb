```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import time
import h5py
import sys  
sys.path.insert(0, '../source/')
from datasets import Interpolated_LCs
from recurrent_models import GRU1D
from experiment import Experiment
from plot_utils import plot_train_val_acc_loss
from sklearn.metrics import confusion_matrix

results_dir = "experiment_results/"
exp_name = "sample_plasticc_rg"
num_epochs = 100

lc_length = 128
seed = 1772670
torch.manual_seed(seed=seed) 
use_gpu = True
lr = 1e-03
wdc = 0
```

```python
#load dataset
interpolated_dataset_filename = "../data/sample_plasticc/plasticc_sample_sns_rg.h5"
interpolated_dataset = Interpolated_LCs(lc_length, interpolated_dataset_filename)
dataset_length = len(interpolated_dataset)
```

```python
#split into train/validation, validation will be ~ .3
val_length = int(dataset_length/3)
train_length = dataset_length - val_length
train_dataset, val_dataset = torch.utils.data.random_split(interpolated_dataset, [train_length, val_length])
```

```python
#dataset loaders
batch_size = 64
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=batch_size, shuffle=True)
```

```python
#define network parameters
params = {
    "input_shape": train_dataset[0][0].shape,
    "num_output_classes" : 4,
    "hidden_size":100,
    "batch_size":batch_size,
    "attention":"self_attention",
    "da":50,
    "r":1
    }

gru_self_attention = GRU1D(params)

experiment = Experiment(
    network_model = gru_self_attention,
    experiment_name = results_dir+exp_name,
    num_epochs = num_epochs,
    learning_rate = lr,
    weight_decay_coefficient = wdc,
    use_gpu = use_gpu,
    train_data = train_loader,
    val_data = val_loader
)
start_time = time.time()
experiment.run_experiment()
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
#visualize results of the training/validation process
results_dir = "experiment_results/"
exp_name = "sn_000000_m_complete"
num_epochs
plot_train_val_acc_loss(results_dir+exp_name+"/result_outputs/",num_epochs)
```

```python
#cm of test set
import matplotlib.pyplot as plt
def plot_cm(true_targets, predictions):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    cm=confusion_matrix(true_targets,predictions)
    cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    im = ax.imshow(cm, interpolation= 'nearest', cmap=plt.cm.Greens)
    namesx = ["snIa","snIb/c","snIIn","snIIP"]
    namesy = ["snIa"," ","snIb/c"," ","snIIn"," ","snIIP"]
    fmt ='.2f'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]): 
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
           # ax.text(j, i, format(cm[i, j]),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")   
    ax.set_xticklabels([''] + namesx)
    ax.set_yticklabels([''] + namesy)
    ax.set_xlabel("predicted class")
    ax.set_ylabel("true class")
```

```python
test_results = pd.read_csv(results_dir+exp_name+"/result_outputs/test_results.csv")
plot_cm(test_results["true_tags"],test_results["predicted_tags"])
```

```python
#which ones where incorreclty classified?
wrong = test_results[test_results.true_tags != test_results.predicted_tags]
how_certain = wrong.predicted_tags.astype(str)
wrong.lookup(how_certain.index, how_certain.values)
```
