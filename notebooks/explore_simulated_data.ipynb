```python
#how many of each type?
def count_sns(snstr):
    sncount = 0
    for i in range(10):
        snstri = snstr.format(i)
        sns = pd.read_pickle(snstri)['lcs']
        sncount += len(sns)
    return sncount
    
snIastr = "../data/training/raw/lcs_Ia_salt2_00000{}.pkl"
snIbcstr = "../data/training/raw/lcs_Ibc_nugent_00000{}.pkl"
snIInstr = "../data/training/raw/lcs_IIn_nugent_00000{}.pkl"
snIIpstr = "../data/training/raw/lcs_IIP_nugent_00000{}.pkl"
```

```python
snIacount = count_sns(snIastr)
print(snIacount)
```

```python
snIbccount = count_sns(snIbcstr)
print(snIbccount)
```

```python
snIIncount = count_sns(snIInstr)
print(snIIncount)
```

```python
snIIpcount = count_sns(snIIpstr)
print(snIIpcount)
```
